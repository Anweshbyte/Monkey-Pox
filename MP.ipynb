```python
import numpy as np
import keras
from PIL import Image
import glob
import tensorflow as tf
```

```python
train = glob.glob("C:/Users/anwes/Downloads/archive/Augmented Images/Augmented Images/*/**")
```

```python
from random import shuffle
shuffle(train)
```

```python
train
```

```python
def index(file_name): 
    if((file_name.find("Monkeypox"))!=-1):
        return 1 
    
    elif((file_name.find("Others"))!=-1):
        return 2
    
    else:
        return 0
```

```python
X_train = np.zeros((len(train),128,128))
Y_train = np.zeros((len(train)))

for i,fl in enumerate(train):
    img = Image.open(fl)
    
    img = img.resize((128,128))
    img = img.convert("L")
    X_train[i,:,:] = np.array(img)
    Y_train[i] = index(fl)
```

```python
len(X_train)
```

```python
model = keras.Sequential([
    
    keras.layers.Conv2D(
    filters=120,
    kernel_size=3,
        # filtering out negative values
    activation='relu',
    input_shape=(128,128,1)
    ),

    keras.layers.Conv2D(
    filters=24,
    kernel_size=1,
    activation='relu',
    ),

    # 1d matrix
    keras.layers.Flatten(),
    # 4
    # reduction layers
    keras.layers.Dense(
        units=96,
        activation='relu',
    ),

    # 96 units -> 4 units
    keras.layers.Dense(4,activation='softmax') 
    
])
```

```python
model.compile(optimizer=keras.optimizers.Adam(learning_rate=1e-3),
              loss='sparse_categorical_crossentropy',
             metrics=['accuracy'])
model.summary()
```

```python
X_train=X_train.reshape((len(train),128,128,1))
model.fit(X_train,Y_train,epochs=7,validation_split=0.1)
```
